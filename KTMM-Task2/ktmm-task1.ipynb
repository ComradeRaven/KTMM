```python
from sympy import Symbol, solve, lambdify, Matrix
import numpy as np
import matplotlib.pyplot as plt
```

### Описание модели 5

Система уравнений:

$$ \begin{cases}
\frac{dx}{dy} = k_{1}z - k_{-1}x - k_{3}xz + k_{-3}y - k_{2}z^{2}x\\
\frac{dx}{dy} = k_{3}xz - k_{-3}y\\
0<=x<=1, 0<=y<=1, 0<=x + 2y<=1\\
\end{cases} $$
    

$k_{1} = 0.12, 
k_{-1} = 0.01, 
k_{2} = 0.95, 
k_{3} = 0.0032, 
k_{-3} = 0.002$

```python
k1_val, k2_val, k3_val = 0.12, 0.95, 0.0032
km1_val, km3_val = 0.01, 0.002
```

```python
# Params
k1, k2, k3 = Symbol("k1"), Symbol("k2"), Symbol("k3")
km1, km3 = Symbol("km1"), Symbol("km3")
# Searched functions
x, y = Symbol("x"), Symbol("y")
# Non-linear component
z = 1 - x - 2*y
```

Решение уравнения относительно $x$ и $k_{1}$

```python
eq1 = k1*z - km1*x - k3*x*z + km3*y - k2*(z**2)*x
eq2 = k3*x*z - km3*y
res = solve([eq1, eq2], y, k1)
xSolution = res[0][0]
k1Solution = res[0][1]
```

Матрица системы, её Якобиан, след, определитель (символьные переменные)

```python
A = Matrix([eq1, eq2])
var_vector = Matrix([x, y])
jacA = A.jacobian(var_vector)
traceA = jacA.trace()
detA = jacA.det()
```

### Однопараметрический анализ

```python
X = np.arange(0, 1, 1e-5)
```

Графики

```python
# Figure
figure = plt.figure(figsize=(20, 10))

# Subplots
subplot = figure.add_subplot()

# Functions to get y and k1 values
y_from_x = lambdify((x, k2, k3, km1, km3), xSolution)
k1_from_x = lambdify((x, k2, k3, km1, km3), k1Solution)

# Plotting
subplot.plot(k1_from_x(X, k2_val, k3_val, km1_val, km3_val), X)
subplot.plot(k1_from_x(X, k2_val, k3_val, km1_val, km3_val), y_from_x(X, k2_val, k3_val, km1_val, km3_val))
```
